```python
logging.basicConfig(
    filename='log_file.log',
    level=logging.INFO,
    format='[%(asctime)s] {%(pathname)s:%(lineno)d} %(levelname)s - %(message)s',
    datefmt='%H:%M:%S',
    filemode='a'
)

class SummaryFormatter(logging.Formatter):
    def format(self, record):
        if record.levelname == 'ERROR':
            return f"{record.levelname}: {record.name}: {record.message[:50]}"  # Summary of error
        return super().format(record)

console = logging.StreamHandler()
console.setLevel(logging.INFO)

formatter = logging.Formatter('%(name)-12s%(levelname)-8s')
summaryFormatter = SummaryFormatter()
console.setFormatter(summaryFormatter)

logging.getLogger('').addHandler(console)

logger = logging.getLogger(__name__)
```

```python
class CustomLogger:
    def __init__(self, name: str, log_gile: str):
        self.logger = logging.getLogger(name=name)
        self.logger.setLevel(logging.DEBUG)
        
        file_handler = logging.FileHandler(log_gile)
        file_handler.setLevel(logging.DEBUG)

        console_handler = logging.StreamHandler()
        console_handler.setLevel(logging.INFO)

        formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')

        file_handler.setFormatter(formatter)
        console_handler.setFormatter(formatter)

        if not self.logger.handlers:
            self.logger.addHandler(file_handler)
            self.logger.addHandler(console_handler)
    

    def debug(self, msg: str):
        self.logger.debug(msg=msg)
    
    def info(self, msg: str):
        self.logger.info(msg=msg)
    
    def warning(self, msg: str):
        self.logger.warning(msg=msg)
    
    def error(self, msg: str):
        self.logger.error(msg=msg)
    
    def critical(self, msg: str):
        self.logger.critical(msg=msg)

```

```python
def drop_columns(input_file: str, columns: list[str], is_xlsx=False) -> pd.DataFrame:
    try:
        if is_xlsx:
            df = pd.read_excel(input_file)
        else:
            df = pd.read_csv(input_file)
        
        logging.info(msg=f'{input_file} successfully saved in pandas dataframe.')

        df = df[columns]
        df.dopna(axis=0, inplace=True)
        df.drop_duplicates(inplace=True)
        return df

    except Exception as e:
        logging.warning(msg=f'Exception --""{e}""-- occured and couldnt get the data.')
        return None
```

```python
class Database:
    def __init__(self):
        self.connection_parameters = {
            key: value for key, value in yaml.load(open('/home/hajali/Desktop/Bargh_Ml_project/configs/database.yaml'), Loader=yaml.SafeLoader).items()
            }
        self.connection = None
    
    def connect(self):
        if self.connection is None:
            self.connection = psycopg2.connect(**self.connection_parameters)
    
    def close(self):
        if self.connection:
            self.connection.close()
            self.connection = None
    
    def execute(self, query:str, params=None):
        with self.connection.cursor() as cursor:
            cursor.execute(query, params)
            return cursor.fetchall()
    
    def commit(self):
        if self.connection:
            self.connection.commit()
    
    def rollback(self):
        if self.connection:
            self.connection.rollback()
    
    def __enter__(self):
        self.connect()
        return self
    
    def __exit__(self):
        self.close()
```

```python
def get_innermost_dict(nested_dict):

    if not isinstance(nested_dict, dict):
        return None  # Not a dictionary

    if not nested_dict:
        return None  # Empty dictionary

    first_value = next(iter(nested_dict.values()))  # Get the first value

    if isinstance(first_value, dict):
        # Recursively call the function if the first value is a dictionary
        return get_innermost_dict(first_value)
    else:
        # If the first value is not a dictionary, then the current dictionary
        # is the innermost dictionary.
        return nested_dict

inner = get_innermost_dict(p)
print(type(inner))
```
