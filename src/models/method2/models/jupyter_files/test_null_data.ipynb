```python
def set_minus(df1, df2, column=None):
    if column is None:
        column = df1.columns.to_list()
    merged = df1.merge(df2, on=column, how='left', indicator=True)
    result = merged[merged['_merge'] == 'left_only'].drop('_merge', axis=1)
    return result
```
