```python
def draw_gen_temp_plot(df_c, x_col, y_col, name, code, color_map, save=False, project_root=None, title="",num=""):
    fig = go.Figure()
    df_show = df_c[(df_c["name"] == name) & (df_c["code"] == code)]
    
    for peak_value, color in color_map.items():
        df_subset = df_show[df_show["is_good_peak"] == peak_value]
        fig.add_trace(go.Scatter(
            x=df_subset[x_col],
            y=df_subset[y_col],
            mode="markers",
            marker=dict(size=4, color=color),
            name=f"is_good_peak = {peak_value}",
            hovertext=df_subset["datetime"]
        ))

    fig.update_traces(marker=dict(size=4, sizemode="diameter", sizeref=1, opacity=0.7))

    if save and project_root:
        save_path = f"{project_root}/src/visualization/unit_figs/report/{num}-{name}-{code}_{title}.html"
        os.makedirs(os.path.dirname(save_path), exist_ok=True)
        fig.write_html(save_path)
    else:
        fig.show()
```

```python
def get_coefs(df_factors):
    df_factors["Date"] = pd.to_datetime(df_factors["Date"])
    coefs = {}
    grouped = df_factors.groupby(['PowerPlantCode', 'PowerPlantName', "UnitCode"])
    for (pp_code, pp_name, unit_code), g in grouped:
        latest_row = g.sort_values("Date", ascending=False).iloc[0]
        coefs[(pp_name, unit_code)] = (latest_row["a1IndexGas"], latest_row["b1IndexGas"])
    return coefs

csv_read_path = os.path.join(project_root, "data", "interim", "factors.csv")
df_factors = pd.read_csv(csv_read_path)
coefs = get_coefs(df_factors)

color_map = {
    0: "purple",
    1: "orange",
    2: "pink",
    3: "blue",
    4: "black",
    5: "red",
    6: "green"
}
name = "قم"
code = "G11"
spliter = (-0.689,119.66736878408216)
draw_gen_temp_plot(df, x_col="temperature", y_col="generation", name=name, code=code, color_map=color_map, save=True,
                   project_root=project_root, title="two-trend-temp",num="9",coefs=coefs)
```
